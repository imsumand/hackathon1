# Problem 1 : Write a python function which will receive 3 input i.e. batch name, start date and end date and returns the leaderboard(in that date window) of that particular batch with the following pieces of info - Name, Marks(till now)

Note: If no date range is provided, show the leaderboard for the entire duration of the course

Marks -10

```python
#there is nothing about end_date
def worsttimetotakehackathon(name,date):
    old=users.merge(batch)
    another=evaluation.groupby('user_id').sum()
    again=old.merge(another,left_on='user_id',right_on='user_id')
    s=again['batch_name']==name
    g=again['batch_start_date']==date
    o=s & g
    b=again[o]
    u=gyan.groupby('user_id').count()
    gy=u['link'].reset_index()
    i=myday.groupby('user_id').count()
    my=i['entry'].reset_index()
    gy1=gy.merge(my)
    gy1.merge(my)
    every=b.merge(gy1)
    x=every['link']+every['entry']+every['score']
    every['total_score']=x
    finished=every.set_index('total_score').sort_index(ascending=False)
    allset=finished['fname'].reset_index()
    return allset
```

# Problem 3: Write a python function that will take batch name and a date range as input and outputs the most consistent student in that given daterange and batch

Note: Come up with your definition of consistency

Marks - 20

```python
def thanksforextension(name,daterange):
    old=users.merge(batch)
    another=evaluation.groupby('user_id').sum()
    again=old.merge(another,left_on='user_id',right_on='user_id')
    s=again['batch_name']==name
    cool=again[s]
    x8=gyan['submission_date']<=daterange
    a=gyan[x8]
    col1=a.groupby('user_id').count()
    x9=myday['entry_date']<=daterange
    a2=myday[x9]
    col2=a2.groupby('user_id').count()
    gy=col1['link'].reset_index()
    every1=cool.merge(gy)
    my=col2['entry'].reset_index()
    every2=every1.merge(my)
    x7=evaluation['evaluation_date']<=daterange
    a3=evaluation[x7]
    col3=a3.groupby('user_id').sum()
    scoring=col3['score'].reset_index()
    scoring.rename(columns={'score':'score_till_now'},inplace=True)
    final=every2.merge(scoring)
    x3=final['score_till_now']/2
    y1=final['link']+final['entry']+x3
    calcu=y1*5
    z=final['late']
    z1=10*z
    ultimate=calcu-z1
    final['most_consistent']=ultimate
    getinn=final.set_index('most_consistent').sort_index(ascending=False)
    okay=getinn.drop_duplicates('batch_id')
    atlast=okay['fname'].reset_index()
    return atlast
    #logic: (score/2+myday+gyan)*5 - 10*(late_submission)
```

```python
thanksforextension("Machine Learning Kolkata 2","2020-02-12")
```

# Problem 2: Write a python function that receives 1 input, batch_name and returns the list of top 10 students who require maximum attention from the teacher.

Note: I know this is kind of abstract, but truly we are facing this problem. Try to come up with a solution/algorithm.

Tip : Don't just think in terms of marks, also think about things like late submissions, number of submissions, myday etc

Tricky scenario: Aisa bhi ho sakta hai koi student bas shuru me acha perform nai kar raha tha, but ab kar raha hai. So consider dates also.

Marks : 50

```python
def attention(id):
    old=users.merge(batch)
    another=evaluation.groupby('user_id').sum()
    again=old.merge(another,left_on='user_id',right_on='user_id')
    s=again['batch_id']==id
    cool=again[s]
    x8=gyan['submission_date']>='2020-02-01'
    a=gyan[x8]
    col1=a.groupby('user_id').count()
    x9=myday['entry_date']>='2020-02-01'
    a2=myday[x9]
    col2=a2.groupby('user_id').count()
    gy=col1['link'].reset_index()
    every1=cool.merge(gy)
    my=col2['entry'].reset_index()
    every2=every1.merge(my)
    x7=evaluation['evaluation_date']>='2020-02-01'
    a3=evaluation[x7]
    col3=a3.groupby('user_id').sum()
    scoring=col3['score'].reset_index()
    scoring.rename(columns={'score':'score_after_today'},inplace=True)
    final=every2.merge(scoring)
    x1=final['score_after_today']/2
    y1=final['link']+final['entry']+x1
    calcu=y1*5
    z=final['late']
    z1=10*z
    ultimate=calcu-z1
    final['requires_attention']=ultimate
    yeah=final.set_index('requires_attention').sort_index()
    atlast=yeah.head(10)
    return atlast

#logic: (score/2+myday+gyan)*5 - 10*(late_submission)
```
